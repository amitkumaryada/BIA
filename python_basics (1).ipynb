```python
def head_count(n):
  """
  this function expect n as input which number of chairs in the room
  and provide output as number of headcount in the room
  """
  count = 0
  occupied_chair = [0,5,6,8,9,11,13,15]
  for i in range(n):
    if i in occupied_chair:
      count = count+1

  return count

```

```python
# Explanation

# The user deposits a fixed amount every month.

# Interest is applied monthly.

# Annual interest rate → convert to monthly rate:
# monthly_rate = annual_rate / 12 / 100

# Each month:

# Add deposit.

# Apply interest to the total.
# Question 1: Monthly Savings with Interest

monthly_deposit = float(input("Enter monthly deposit amount: "))
annual_interest_rate = float(input("Enter annual interest rate (in %): "))
months = int(input("Enter number of months: "))

monthly_rate = annual_interest_rate / 12 / 100
savings = 0

for _ in range(months):
    savings += monthly_deposit           # Add deposit
    savings += savings * monthly_rate    # Apply interest

print(f"\nTotal savings after {months} months: ${savings:.2f}")
```

```python
# Question 2: Debt Repayment Plan (Using a While Loop)
# Explanation

# The user starts with an initial debt.

# Monthly interest accumulates.

# Each month:

# Add interest to debt.

# Subtract monthly payment.

# Loop continues until debt ≤ 0.

# Question 2: Debt Repayment Plan

debt = float(input("Enter initial debt amount: "))
monthly_payment = float(input("Enter monthly payment: "))
annual_interest_rate = float(input("Enter annual interest rate (in %): "))

monthly_rate = annual_interest_rate / 12 / 100
months = 0

while debt > 0:
    debt += debt * monthly_rate     # Add monthly interest
    debt -= monthly_payment         # Make payment
    months += 1

print(f"\nDebt paid off in {months} months.")
```

```python
def amortization_schedule(loan_amount, annual_rate, years):
    monthly_rate = annual_rate / 12 / 100
    n_payments = years * 12

    # Monthly payment using amortization formula
    monthly_payment = loan_amount * (monthly_rate * (1 + monthly_rate)**n_payments) / ((1 + monthly_rate)**n_payments - 1)

    balance = loan_amount
    schedule = []

    for month in range(1, n_payments + 1):
        interest = balance * monthly_rate
        principal = monthly_payment - interest
        balance -= principal
        schedule.append((month, round(monthly_payment, 2), round(principal, 2), round(interest, 2), round(balance, 2)))

    return schedule


# Use the function for $100,000 loan, 5% annual interest, 15 years
schedule = amortization_schedule(100000, 5, 15)

# Print schedule
print("Month | Payment | Principal | Interest | Balance")
for row in schedule[:12]:  # printing only first 12 months for readability
    print(row)

print("\nFinal Balance:", schedule[-1][-1])
```

```python
def analyze_budget(budget, expenses):
    total_expenses = sum(expenses)
    remaining = budget - total_expenses

    if remaining > 0:
        status = "Within Budget"
    elif remaining < 0:
        status = "Over Budget"
    else:
        status = "Exactly on Budget"

    return {
        "Budget": budget,
        "Total Expenses": total_expenses,
        "Remaining": remaining,
        "Status": status
    }


result = analyze_budget(1000, [300, 400, 200])
print(result)
```

```python

# Drop irrelevant columns (RowNumber, CustomerId, Surname)
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Pairplot for selected features
sns.pairplot(df[['CreditScore', 'Age', 'Balance', 'NumOfProducts', 'Exited']], hue='Exited', markers=['o', 's'])
plt.suptitle('Pairplot of Selected Features by Churn Status', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Exited', multiple='stack', kde=True, palette='viridis')
plt.title('Age Distribution by Churn Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation Matrix Heatmap
correlation_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix - Bank Churn Data')
plt.show()
```

```python
# Bar Chart: Churn by Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', hue='Exited', data=df, palette='pastel')
plt.title('Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot: Balance by Churn Status
plt.figure(figsize=(8, 6))
sns.boxplot(x='Exited', y='Balance', data=df, palette='Set2')
plt.title('Balance Distribution by Churn Status')
plt.xlabel('Churn Status')
plt.ylabel('Balance')
plt.show()
```

```python
# Violin Plot: Age by Churn
plt.figure(figsize=(8, 6))
sns.violinplot(x='Exited', y='Age', data=df, palette='Set3')
plt.title('Age Distribution by Churn Status')
plt.xlabel('Churn Status')
plt.ylabel('Age')
plt.show()
```
